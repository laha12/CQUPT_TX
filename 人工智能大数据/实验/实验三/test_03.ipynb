```python
## 实例一
import pymysql
from openpyxl import load_workbook
# 创建数据库
db = pymysql.connect(host='localhost',
    user='root',
    password='123456',
    database='student_510')
cursor = db.cursor() 
cursor.execute("DROP TABLE IF EXISTS mingdan") 
sql = """
    CREATE TABLE mingdan
    (
    学号 VARCHAR(10) PRIMARY KEY,
    姓名 VARCHAR(10)
    )
    """
cursor.execute(sql) 
print('表结构创建成功！') 
# 表格输出
wb = load_workbook("student_101.xlsx")
ws = wb.worksheets[0] #ws = wb.worksheets["student"]
dict_stu={}
for row in ws:
    dict_stu[row[0].value]=row[1].value
wb.close() 
del dict_stu['学号'] #dict_stu.pop("学号")
print(dict_stu)
# inster插入到数据库
for xh , xm in dict_stu.items():
    if xh<10:
        xh="0"+str(xh)
    else:
        xh=str(xh)
    sql="INSERT INTO mingdan VALUES('"+ xh + "','" + xm + "'" +")"
    try:
        cursor.execute(sql) 
        db.commit()
    except:
        db.rollback() 
        print('记录插入失败！')
print("%d个学生名单插入成功！"%len(dict_stu))
cursor.close()
db.close() 
```

```python
## 思考一
import pymysql
import pandas as pd
# 从test01数据库读取并转字典
db_pymysql = pymysql.connect(host='localhost',
    user='root',
    password='123456',
    database='test01')
df = pd.read_sql('select * from student01',con=db_pymysql) #打开数据库“test01”，以 DataFrame
nrows=df.shape[0] #获取行数
ncols=df.shape[1] #获取列数
dic_score={}
for r in range(0,nrows): #逐行逐列遍历
    list_stu=[]
    for c in range(1,ncols):
        list_stu.append(df.iloc[r,c])
    dic_score[df.iloc[r,0]]=list_stu
print(dic_score)

# 统计每名学生的平均成绩和最高成绩，添加至每个学生成绩的后面
for name, scores in dic_score.items():
    average_score = sum(scores) / len(scores)
    max_score = max(scores)
    dic_score[name].extend([average_score, max_score])
# 求每门课的最高分，作为新的键和值存入字典
MaxScore=[0,0,0,0,0,0]
for scores in dic_score.values():
    MaxScore[0] = max(MaxScore[0], scores[0])
    MaxScore[1] = max(MaxScore[1], scores[1])
    MaxScore[2] = max(MaxScore[2], scores[2])
    MaxScore[3] = max(MaxScore[3], scores[3])
    MaxScore[4] = max(MaxScore[4], scores[4])
    MaxScore[5] = max(MaxScore[5], scores[5])
dic_score["最高分"]=MaxScore
print("最终成绩表：")
print("姓名\t语文\t数学\t英语\t计算机\t平均分\t最高分")
for name, scores in dic_score.items():
    print(f"{name}\t{scores[0]}\t{scores[1]}\t{scores[2]}\t{scores[3]}\t{scores[4]:.2f}\t{scores[5]}")
# 连接test02数据库 并创建表
db = pymysql.connect(host='localhost',
    user='root',
    password='123456',
    database='test02')
cursor = db.cursor() 
cursor.execute("DROP TABLE IF EXISTS student02") 
sql = """
CREATE TABLE student02
    (
    姓名 VARCHAR(10),
    语文 INT,
    数学 INT,
    英语 INT,
    计算机 INT,
    平均分 INT,
    最高分 INT
    )
    """
cursor.execute(sql) 
print('表结构创建成功！') 
cursor.close()
db.close()
# 数据导入test02
db = pymysql.connect(host='localhost',
    user='root',
    password='123456',
    database='test02')
cursor = db.cursor()
# inster插入到数据库
for xh , xm in dic_score.items():
    sql="INSERT INTO student02 VALUES('"+ xh + "',"+str(xm[0])+","+str(xm[1])+","+str(xm[2])+","+str(xm[3])+","+str(xm[4])+","+str(xm[5])+")"
    try:
        cursor.execute(sql) 
        db.commit()
    except:
        db.rollback() 
        print('记录插入失败！')
print('插入成功')
cursor.close()
db.close()
```
