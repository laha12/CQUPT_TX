```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, silhouette_samples
N_CLUSTERS = 7                                     # 类簇的数量
MARKERS = ['*', 'v', '+', '^', 's', 'x', 'o']      # 标记样式（绘图）
COLORS = ['r', 'g', 'm', 'c', 'y', 'b', 'orange']  # 标记颜色（绘图）
DATA_PATH = 'China_cities.csv'              # 数据集路径
df = pd.read_csv(DATA_PATH)
df.to_excel('Cites_510.xlsx', index=False)
x = df.drop('省级行政区', axis=1)
x = x.drop("城市", axis=1)
x_np = np.array(x)        # 将x转化为numpy数组
model = KMeans(N_CLUSTERS)      # 构建聚类器
model.fit(x)                    # 训练聚类器
labels = model.labels_      # 获取聚类标签
plt.figure(figsize=(9, 6))
plt.title("Major Cities in China", fontsize=22)
plt.xlabel('East Longitude', fontsize=18)
plt.ylabel('North Latitude', fontsize=18)

for i in range(N_CLUSTERS):
    members = labels == i      # members是一个布尔型数组
    plt.scatter(
        x_np[members, 1],      # 城市经度数组
        x_np[members, 0],      # 城市纬度数组
        marker = MARKERS[i],   # 标记样式
        c = COLORS[i]          # 标记颜色
    )   # 绘制散点图
plt.savefig('Cities.png')
plt.grid()
plt.show()
```

```python
## 思考一
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans            #导入 K-Means 聚类算法模型
import matplotlib.pyplot as plt
path='./iris.data' #数据文件路径，与代码文件在同一个文件夹
data=pd.read_csv(path)
data.to_excel('flower.xlsx', index=False)
col_1=data["花萼长度"]
data_1=np.array(col_1)
col_2=data["花萼宽度"]
data_2=np.array(col_2)
col_3=data["花瓣长度"]
data_3=np.array(col_3)
col_4=data["花瓣宽度"]
data_4=np.array(col_4)
col_5=data["种类"]
data_5=np.array(col_5)
s="{0}\t{1:^8}\t{2:^8}\t{3:^8}\t{4:^8}\t{5:^8}"
print("序号\t花萼长度\t花萼宽度\t花瓣长度\t花瓣宽度\t种类")
for i in range(0,10):
    print(s.format(i,data_1[i],data_2[i],data_3[i],data_4[i],data_5[i]))
    print("\n\n")
X=[]
for i in range(0, len(data)):           #print (data.iloc[i]['花萼长度'], data.iloc[i]['花萼宽度'])
    X.append([data.iloc[i]['花萼长度'], data.iloc[i]['花萼宽度']])
X=np.array(X) #转换成 numpy array(数组)
plt.rcParams['font.sans-serif']=['SimHei']     #用来正常显示中文标签
plt.figure(1,figsize=(8,4))                   #定义图表大小
n=3                                            #类簇的个数k
km=KMeans(n_clusters=n)
km.fit(X)                                     #把数据和对应的分类数据放入聚类函数中进行聚类
km.labels_                                    #X 中每项所属分类的一个列表
markers = ['*', 'v', '+', '^', 's', 'x', 'o']                     #图中显示的标识
for i in range(n):                           #循环 3 次（共 3 个类簇），画出 15 个点
    m=(km.labels_==i)
    plt.scatter(x=X[m,0],y=X[m,1],s=100,marker=markers[i],c='b',alpha=0.9)
                        #x,y 为输入数据，s 为点大小；c 为点颜色；marker 为点形状；alpha 为透明度（0～1）
plt.xlabel("花萼长度",fontsize = 20,color = 'green')                              #X 轴标签(标题)
plt.ylabel("花萼宽度",fontsize = 20,color = 'magenta')                            #Y 轴标签
plt.title('鸢尾花按花萼聚类-510-雷宇航 ',family = "SimHei",fontsize = 20,color = 'red') #标题
plt.xticks(fontsize=16)                       #x轴刻度值字号
plt.yticks(fontsize=16)                      #y轴刻度值字号
plt.savefig('Flower.png', dpi=300, bbox_inches='tight')
plt.show()                                    #显示图形
```
