```python
## 实例五
from itertools import chain 
from pypinyin import pinyin, Style #导入拼音库 用于对中文字符排序
def to_pinyin(s): 
    return ''.join(chain.from_iterable(pinyin(s, style=Style.TONE3)))
city = input("请输入5个城市的名字，用空格分隔：") 
City = city.split(" ",4) 
print('for循环遍历输出：')#利用for循环输出城市名 
for i in City: 
    print('{}'.format(i),end=" ") 
print('\n索引值输出：{}'.format(City[0])) 
print('切片输出：',City[1:5]) 
print('逆序输出：',end="")#各种方法实现逆序输出 
print('切片逆序：',end="") 
print(City[::-1]) 
cit2 = City.copy() # reverse是就地操作 会改变原列表 所以需要复制一份
cit2.reverse() 
print('reverse逆序：',end="") 
print(cit2) 
cit3 = City.copy() 
print('sort：',end="")#两种方法实现排序 sort就地 sorted是新列表
cit3.sort(reverse=True,key=to_pinyin)  #利用reverse实现降序
print(cit3) 
cit4 = City.copy() 
print('sorted：',end="") 
print(sorted(cit4,reverse=True,key=to_pinyin)) 
a = cit3.index('重庆')#利用index函数查询下标 
print("我的家乡城市：",cit3[a]) 
l = len(cit3) #城市列表长度
#输出家乡前后的城市 
if (a-1) >= 0: 
    print('前面的城市是：',cit3[0:a]) 
if (a+1) <= (l-1): 
    print('后面的城市是：',cit3[a+1:l]) 
#修改前后城市名
if (a-1) >= 0:
    cit3[a-1] = input("请输入城市名：")
    print("前面修改完成！") 
else:
    print('前面无其他城市')
if (a+1) <= (l-1): 
    cit3[a+1] = input("请输入城市名：")
    print("后面修改完成") 
else:
    print('后面无其他城市')
print(cit3) 
#删除城市并输出 
if (a-1) >= 0:
    cit3.pop(a-1)
    if(a+1)<=len(cit3): # pop默认弹出一个 总长度减一
        cit3.pop(a)
else:
    if (a+1) <= (l-1): 
        cit3.pop(a+1) 
print("剩余的城市为：",cit3,end="")
```

```python
dic_score = {
    "张三": [88, 90, 98, 95],
    "李四": [85, 92, 95, 98],
    "王五": [89, 89, 90, 92],
    "丁六": [82, 86, 89, 90]
}

# (1) 统计每名学生的平均成绩和最高成绩，添加至每个学生成绩的后面，最后将字典输出。
for name, scores in dic_score.items():
    average_score = sum(scores) / len(scores)
    max_score = max(scores)
    dic_score[name].extend([average_score, max_score])

print("（1）修改后学生成绩：")
print(dic_score)

# (2) 将（1）中的成绩形成嵌套列表输出。
nested_list = [[name] + scores for name, scores in dic_score.items()]
print("（2）嵌套列表学生成绩：")
print(nested_list)

# (3) 求每门课的最高分，作为新的键和值存入字典，并按照如下格式输出。
MaxScore=[0,0,0,0,0,0]
for scores in dic_score.values():
    MaxScore[0] = max(MaxScore[0], scores[0])
    MaxScore[1] = max(MaxScore[1], scores[1])
    MaxScore[2] = max(MaxScore[2], scores[2])
    MaxScore[3] = max(MaxScore[3], scores[3])
    MaxScore[4] = max(MaxScore[4], scores[4])
    MaxScore[5] = max(MaxScore[5], scores[5])
dic_score["最高分"]=MaxScore
print("（3）最终成绩表：")
print("姓名\t语文\t数学\t英语\t计算机\t平均分\t最高分")
for name, scores in dic_score.items():
    print(f"{name}\t{scores[0]}\t{scores[1]}\t{scores[2]}\t{scores[3]}\t{scores[4]:.2f}\t{scores[5]}")
```
